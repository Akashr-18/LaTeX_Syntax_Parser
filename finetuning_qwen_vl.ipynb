```python
from unsloth import FastVisionModel
import torch
```

```python
fourbit_models = [
    "unsloth/Llama-3.2-11B-Vision-Instruct-bnb-4bit",
    "unsloth/Qwen2-VL-7B-Instruct-bnb-4bit"
]
```

```python
model, tokenizer = FastVisionModel.from_pretrained(
    "unsloth/Qwen2-VL-7B-Instruct",
    load_in_4bit=True, #Compression technique to reduce the memory of the model(from 32 or 16 bits) to 4 bits
    use_gradient_checkpointing="unsloth"
)
```

```python
model = FastVisionModel.get_peft_model(
    model,
    finetune_vision_layers=True,
    finetune_language_layers=True,
    finetune_attention_modules=True,
    finetune_mlp_modules=True,

    r=16,  #Rank of LORA Adapter
    lora_alpha=16,  #Scaling factor for Lora adapters
    lora_dropout=0,
    bias="none",
    random_state = 3407,
    use_rslora=False,  #Rank stabilized Lora
    loftq_config=None  #Disables Lora loft config quantization technique
)
```

```python
instruction = "Write the LaTex representation for this image."
```

```python
def convert_to_conversation(sample):
  conversation = [
      {"role": "user",
       "content": [
           {"type": "text", "text": instruction},
           {"type": "image", "image": sample["image"]}
       ]
       },
      {"role": "assistant",
       "content": [
           {"type": "text", "text": sample["text"]}
       ]
       }
  ]
  return {"messages": conversation}
```

```python
convert_to_conversation(dataset[0])
```

```python
converted_dataset = [convert_to_conversation(sample) for sample in dataset]
```

```python
converted_dataset[1]
```

```python
image = dataset[1]["image"]
messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": instruction},
            {"type": "image", "image": image}
        ]
    }
]
```

```python
input_text = tokenizer.apply_chat_template(messages, add_generation_prompt=True) #Adds the special tokens that signal the model to generate a response for qwen models
inputs = tokenizer(
    image, input_text,
    add_special_tokens = False,
    return_tensors = "pt",
).to("cuda")
```

```python
from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer, skip_prompt=True)
_ = model.generate(**inputs, streamer= text_streamer, max_new_tokens = 128, use_cache=True, temperature=1.5, min_p=0.1) #min_p -> minimum probability token
```

```python
from unsloth import is_bf16_supported
from unsloth.trainer import UnslothVisionDataCollator
from trl import SFTTrainer, SFTConfig #Supervised Finetuning. Q n A
```

```python
FastVisionModel.for_inference(model)
```

```python
trainer = SFTTrainer(
    model = model,
    tokenizer=tokenizer,
    data_collator = UnslothVisionDataCollator(model, tokenizer), #Data Collator is to ensure the batches of union len for img and text
    train_dataset = converted_dataset,
    args = SFTConfig(
        per_device_train_batch_size=2,  # 2 samples simulataneuosly
        gradient_accumulation_steps=4,  # Updates weight after accumulating gradients from 4 batches (After 8 images)
        warmup_steps = 5,  #No of steps before training starts
        max_steps=30,
        learning_rate=2e-4,
        fp16=not is_bf16_supported(),
        bf16 = is_bf16_supported(),
        logging_steps=1,
        optim="adamw_8bit",
        weight_decay=0.01, #Penalizes large weights to improve generalization
        lr_scheduler_type="linear", #After warmup learning rate decreased linearly to 0
        seed=3407,
        output_dir="outputs",
        report_to = "none",
        remove_unused_columns=False,  #IMP. Preserves image data during batch preparation
        dataset_text_field="",
        dataset_kwargs = {"skip_prepare_dataset": True},
        dataset_num_proc=4, #Parallel workers
        max_seq_length=2048,
    ),
)
```

```python
trainer.train()
```
